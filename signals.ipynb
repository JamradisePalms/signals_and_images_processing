```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.graph_objects as go
F = 64
```

# Signals

```python
data_sin = pd.read_csv('sin_signal_sampled.csv')
plt.figure()
plt.plot(data_sin.iloc[:, 1])
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()
```

```python
data_noisy_sin = pd.read_csv('noisy_sin_signal.csv')
plt.figure()
plt.plot(data_noisy_sin)
plt.xlabel('Samples')
plt.ylabel('Values')
plt.show()
```

```python
sin_hist = pd.read_csv('histogram_noisy_gaussian_sin_signal.csv')
plt.figure()
plt.bar(range(len(sin_hist)), sin_hist.iloc[:, 1])
plt.xlabel('Bin')
plt.ylabel('Count')
```

```python
data_sin_complex = pd.read_csv('sin_signal_complex.csv')
magnitude_sin = data_sin_complex['magnitude'].values
n_sin = len(magnitude_sin)

frequencies = np.fft.fftshift(np.fft.fftfreq(n_sin, d=1 / F))

rearranged_magnitude = np.fft.fftshift(magnitude_sin)

fig = go.Figure()

fig.add_trace(go.Bar(
    x=frequencies,
    y=rearranged_magnitude,
    name='Magnitude Spectrum'
))

fig.update_layout(
    title='Corrected Magnitude Spectrum for Sine Signal',
    xaxis_title='Frequency (Hz)',
    yaxis_title='Magnitude',
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    template='plotly_white'
)

fig.show()
```

```python
data_double_square_complex = pd.read_csv('double_square_signal_complex.csv')
magnitude_double_square = data_double_square_complex['magnitude']
n_double_square = len(magnitude_double_square)
half_n_double_square = n_double_square // 2
rearranged_double_square = np.concatenate((magnitude_double_square[half_n_double_square:], magnitude_double_square[:half_n_double_square]))

plt.figure()
plt.bar(range(-half_n_double_square, half_n_double_square), rearranged_double_square)
plt.xlabel('Frequency (index)')
plt.ylabel('Magnitude')
plt.title('Rearranged Magnitude Spectrum for Double Square Signal')
plt.grid()
plt.show()
```
