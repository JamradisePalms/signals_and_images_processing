```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_sin = pd.read_csv('sin_signal_sampled.csv')
plt.figure()
plt.plot(data_sin.iloc[:, 1])
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()
```

```python
data_noisy_sin = pd.read_csv('noisy_sin_signal.csv')
plt.figure()
plt.plot(data_noisy_sin)
plt.xlabel('Samples')
plt.ylabel('Values')
plt.show()
```

```python
data_sin_complex = pd.read_csv('sin_signal_complex.csv')
magnitude_sin = data_sin_complex['magnitude']
n_sin = len(magnitude_sin)
half_n_sin = n_sin // 2
rearranged_sin = np.concatenate((magnitude_sin[half_n_sin:], magnitude_sin[:half_n_sin]))

plt.figure()
plt.bar(range(-half_n_sin, half_n_sin), rearranged_sin)
plt.xlabel('Frequency (index)')
plt.ylabel('Magnitude')
plt.title('Rearranged Magnitude Spectrum for Sine Signal')
plt.grid()
plt.show()
```
