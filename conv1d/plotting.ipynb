```python
import pandas as pd
import plotly.graph_objects as go
def load_signal(file_name):
    data = pd.read_csv(file_name)
    return data['value'].values
```

```python
original_signal_file = "conv1d/original_triangle_signal.csv"
low_passed_signal_file = "conv1d/low_passed_triangle_signal.csv"
high_passed_signal_file = "conv1d/high_passed_triangle_signal.csv"

original_signal = load_signal(original_signal_file)
low_passed_signal = load_signal(low_passed_signal_file)
high_passed_signal = load_signal(high_passed_signal_file)

fig = go.Figure()

fig.add_trace(go.Scatter(
    y=original_signal,
    mode='lines',
    name='Original Signal',
    line=dict(color='blue')
))

fig.add_trace(go.Scatter(
    y=low_passed_signal,
    mode='lines',
    name='Low-pass Filtered',
    line=dict(color='green')
))

fig.add_trace(go.Scatter(
    y=high_passed_signal,
    mode='lines',
    name='High-pass Filtered',
    line=dict(color='red')
))

fig.update_layout(
    title="Interactive Signal Filtering Visualization",
    xaxis_title="Sample Index",
    yaxis_title="Amplitude",
    legend=dict(x=0.01, y=0.99),
    template="plotly_white",
    hovermode="x",
)

fig.update_xaxes(rangeslider_visible=True)
fig.update_yaxes(fixedrange=False)

fig.show()
```
